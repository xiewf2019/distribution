```python
import igraph as ig
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
```

```python
import igraph as ig
import leidenalg as la

G = ig.Graph.Famous('Zachary')
partition = la.find_partition(G, la.ModularityVertexPartition)

```

```python
plt.figure(figsize=(40, 40))

# Generate four different graphs
# Graph 1: Erdős-Rényi graph
g1 = ig.Graph.Erdos_Renyi(n=10, p=0.5)
print(g1)
# Graph 2: Barabási-Albert graph
g2 = ig.Graph.Barabasi(n=10, m=2)
# Graph 3: Watts-Strogatz graph
g3 = ig.Graph.Watts_Strogatz(dim=1, size=10, nei=2, p=0.5)
# Graph 4: Ring graph
g4 = ig.Graph.Ring(n=10)

# Plot the graphs
plt.subplot(221)
ig.plot(partition,bbox={300,300}, target="sb.pdf")
plt.title('Erdős-Rényi Graph')

plt.subplot(222)
ig.plot(g2, target=plt.gca())
plt.title('Barabási-Albert Graph')

plt.subplot(223)
ig.plot(g3, target=plt.gca())
plt.title('Watts-Strogatz Graph')

plt.subplot(224)
ig.plot(g4, target=plt.gca())
plt.title('Ring Graph')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Erdős-Rényi Graph',
           markersize=10, markerfacecolor='skyblue'),
    Line2D([0], [0], marker='o', color='w', label='Barabási-Albert Graph',
           markersize=10, markerfacecolor='lightgreen'),
    Line2D([0], [0], marker='o', color='w', label='Watts-Strogatz Graph',
           markersize=10, markerfacecolor='lightcoral'),
    Line2D([0], [0], marker='o', color='w', label='Complete Graph',
           markersize=10, markerfacecolor='violet'),
]

# Add the legend to the figure
plt.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=4)
plt.suptitle('Four Different Types of Graphs', fontsize=16)

plt.show()
```
